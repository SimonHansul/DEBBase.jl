The NEC function is given by 

$$
y = \alpha \ max(0, D - D_{NEC})
$$

We can convert this to a differentiable function by replacing $max(.)$ with a sigmoid function:


$$

y = \alpha \frac{1}{1 + e^{-\beta (D - D_{NEC})}} (D - D_{NEC})
$$

For a high value of $\beta$, the graphs for the two functions will be equivalent.

```julia
function NEC(D, α, D_NEC)
    return α * max(0, D - D_NEC)
end

function contNEC(D, α, β, D_NEC)
    return α * (1 / (1 + exp(-β * (D - D_NEC)))) * (D - D_NEC)
end

α = 0.5
β = 20.
D_NEC = 1.
plot(x -> NEC(x, α, D_NEC), linestyle = :dash)
plot!(x -> contNEC(x, α, β, D_NEC))
```
